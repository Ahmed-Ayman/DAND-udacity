```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
'''
the final_data file includes the moving average
of the temperatures per each 10 years.
'''

final_data = pd.read_csv('final_data.csv')
xAxis = final_data['year']
cairo_tmp = final_data['avg_temp_cairo']
glabal_tmp = final_data['avg_temp_global']

#plotting
plt.plot(xAxis,cairo_tmp)
plt.plot(xAxis,glabal_tmp)
plt.legend(['y = cairo_tmp', 'y = global_tmp '])
plt.show()
```
